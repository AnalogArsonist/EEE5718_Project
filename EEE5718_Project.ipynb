```python
%matplotlib inline

#Citations:
#Veeresh Taranalli, "CommPy: Digital Communication with Python, version 0.3.0. Available at https://github.com/veeresht/CommPy", 2015.

import numpy as np
import math
import matplotlib.pyplot as plt
import sys
import logging

from bitstring import BitArray, Bits, BitStream, BitString, ConstBitStream, ReadError
from scipy import signal
from matplotlib.ticker import NullFormatter
from scipy.special import erfc
from numpy import complex, sum, pi, arange, array, size, shape, real, sqrt
from numpy import matrix, sqrt, sum, zeros, concatenate, sinc
from numpy.random import randn, seed, random

logging.basicConfig(stream=sys.stderr, level=logging.DEBUG)

class mpeg2_parse(object): ### Parses MPEG2 header
    def __init__(self, s, byte):
        base = (byte-1)*188*8
        self.sync = s[base:base+8]
        self.TEI = s[base+8:base+9]
        self.USI = s[base+9:base+10]
        self.TP = s[base+10:base+11]
        self.PID = s[base+11:base+24]
        self.TSC = s[base+24:base+26]
        self.AFC = s[base+26:base+28]
        self.CC = s[base+28:base+32]
        self.payload = s[base+32:base+188*8] # payload
        self.full = s[base:base+188*8]
        """
        self.sync = s[0:8]
        self.TEI = s[8:9]
        self.USI = s[9:10]
        self.TP = s[10:11]
        self.PID = s[11:24]
        self.TSC = s[24:26]
        self.AFC = s[26:28]
        self.CC = s[28:32]
        self.payload = s[32:188*8] # payload
        self.full = s[0:188*8]
        """
        
def rrcosfilter(N, alpha, Ts, Fs, Norm=True):
    """
    Generates a root raised cosine (RRC) filter (FIR) impulse response.

    Parameters
    ----------
    N : int
        Length of the filter in samples.

    alpha : float
        Roll off factor (Valid values are [0, 1]).

    Ts : float
        Symbol period in seconds.

    Fs : float
        Sampling Rate in Hz.

    Returns
    ---------

    h_rrc : 1-D ndarray of floats
        Impulse response of the root raised cosine filter.

    time_idx : 1-D ndarray of floats
        Array containing the time indices, in seconds, for
        the impulse response.
    """

    T_delta = 1/float(Fs)
    time_idx = ((np.arange(N+1)-N/2))*T_delta
    sample_num = np.arange(N+1)
    h_rrc = np.zeros(N+1, dtype=float)

    for x in sample_num:
        t = (x-N/2)*T_delta
        if t == 0.0:
            h_rrc[x] = 1.0 - alpha + (4*alpha/np.pi)
        elif alpha != 0 and t == Ts/(4*alpha):
            h_rrc[x] = (alpha/np.sqrt(2))*(((1+2/np.pi)* \
                    (np.sin(np.pi/(4*alpha)))) + ((1-2/np.pi)*(np.cos(np.pi/(4*alpha)))))
        elif alpha != 0 and t == -Ts/(4*alpha):
            h_rrc[x] = (alpha/np.sqrt(2))*(((1+2/np.pi)* \
                    (np.sin(np.pi/(4*alpha)))) + ((1-2/np.pi)*(np.cos(np.pi/(4*alpha)))))
        else:
            h_rrc[x] = (np.sin(np.pi*t*(1-alpha)/Ts) +  \
                    4*alpha*(t/Ts)*np.cos(np.pi*t*(1+alpha)/Ts))/ \
                    (np.pi*t*(1-(4*alpha*t/Ts)*(4*alpha*t/Ts))/Ts)


    if Norm == True:
        h_rrc = h_rrc/np.sum(h_rrc)
        
        
    return time_idx, h_rrc

def stream_to_QPSK(s, os):
    """
    Converts a bit stream into I and Q streams, oversamples, and converts to Gray-coded QPSK/4-QAM
    with absolute mapping.
    
    s = BitStream object
    os = oversampling rate
    """
    # split into I and Q streams and convert from BitStream object to strings
    iString = s[0::2].bin
    qString = s[1::2].bin
    
    # over sample I and Q streams as strings -- does not stuff with zeros and interpolate but converts dirac to rectangular waves
    iStringOversampled = np.repeat([iString[i] for i in range(len(iString))],os)
    qStringOversampled = np.repeat([qString[i] for i in range(len(qString))],os)
    
    # The below is an attemp to use the resample function but need to convert string to array first
    # and pass the RRC filter as the 'window', I think.
    #iStringOversampled = signal.resample([iString[i] for i in range(len(iString))],os*len(iString))
    #qStringOversampled = signal.resample([qString[i] for i in range(len(qString))],os*len(qString))
    
    # convert strings to integers and perform QPSK, i.e., 4-QAM shifting values to -1 or +1
    iData = [2*int(iStringOversampled[i])-1 for i in range(len(iStringOversampled))]
    qData = [2*int(qStringOversampled[i])-1 for i in range(len(qStringOversampled))]
    
    return iData, qData

def awgn(input_signal, snr_dB, rate=1.0):
    """
    Addditive White Gaussian Noise (AWGN) Channel.
    Parameters
    ----------
    input_signal : 1D ndarray of floats
        Input signal to the channel.
    snr_dB : float
        Output SNR required in dB.
    rate : float
        Rate of the a FEC code used if any, otherwise 1.
    Returns
    -------
    output_signal : 1D ndarray of floats
        Output signal from the channel with the specified SNR.
    """

    avg_energy = sum(input_signal * input_signal)/len(input_signal)
    snr_linear = 10**(snr_dB/10.0)
    noise_variance = avg_energy/(2*rate*snr_linear)

    if input_signal.dtype is complex:
        noise = sqrt(noise_variance) * randn(len(input_signal)) * (1+1j)
    else:
        noise = sqrt(2*noise_variance) * randn(len(input_signal))

    output_signal = input_signal + noise

    return output_signal

"""
======================================================
Set the design parameters in this section
======================================================
"""

bufferSize = 188*8*8 # input buffer size (bits) -- Each MPEG-2 Transport Stream (TS) packet is 188 bytes

Rs = 10. # Symbol rate - for QPSK it is 1/2 bit rate
Tsymbol = 1/Rs # Symbol duration

span = 6 # length of RCC filter in symbols (must be even)
sample_per_symbol = 10 # samples per symbol for RRC filter function
N = span * sample_per_symbol # length of the RCC filter in samples
alpha = 0.35 # Roll-off factor (excessive bandwidth, i.e., 0.5*Rs*(1+alpha)) for RRC filter (SSB)
# For DSB (RF), the total bandwidth of the filter is just Rs*(1+alpha)

fMod = 70 # carrier modulation frequency
fSample = 10*fMod # sampling rate -- higher the better but should be at least 2x fMod
Tsample = 1./fSample # time step (sampling period)

upsample_rate = fSample/Rs # used to upsample I/Q pulses before convolving with RRC filter
#dataLen = int(bufferSize*0.5*upsample_rate) # length of I and Q arrays after upsampling

plots = True # Turn on plots

SNR_MIN=10
SNR_MAX=100
SNR=SNR_MIN

"""
======================================================
Set up file stream
======================================================
"""

s = BitStream(filename='tst.ts') # create BitStream object

"""
======================================================
Generate RRC Filter Coefficients
======================================================
"""

# total length in time will be bufferSize*0.5*Tsymbol (Oversampling_rate*Tsample)
print('Buffer Size = ', bufferSize)
print('Symbol Period = ', Tsymbol)
print('Sampling Rate (Hz) = ', 1/Tsample)
print('Upsample Rate = ', upsample_rate)

# Generate root-raised cosine (RRC) filter impulse response
time_idx,h_rrc = rrcosfilter(N, alpha, Tsymbol, sample_per_symbol/Tsymbol, Norm=True) # generate time index and impulse response of filter

# Plot RRC Impulse Response
if plots == True:
    plt.stem(time_idx,h_rrc)
    plt.title('Root-Raised Cosine (RRC) Impulse Response')
    plt.show()

# Plot RRC Frequency Response
if plots == True:
    w,h = signal.freqz(h_rrc)
    plt.plot(w,20*np.log10(abs(h))) # Signal amplitude
    plt.title('RRC Frequency Response')
    plt.axis('tight')
    plt.show()
    #print(w)

"""
======================================================
Generate modulation signals
======================================================
"""

x = np.arange(0, Tsymbol*bufferSize/2, Tsample) # time steps for modulation
iMod = np.cos(2 * np.pi * fMod * x)/math.sqrt(2) # modulator for inphase
qMod = np.sin(2 * np.pi * fMod * x)/math.sqrt(2) # modulator for quadrature

"""
======================================================
Start main loop to operate on MPEG-2 files
======================================================
"""

inc=0
try:
    while inc < 1:
        streamBuffer = s.read(bufferSize) # read bits from file

        #frame = mpeg2_parse(streamBuffer) # parse frame into header and payload -- needs to be able to handle >1 frame
        # pass desired frame number to mpeg2_parse class?
        #print(frame.sync.hex)
        
        """
        ======================================================
        Mux-Adapation
        ======================================================
        """
        streamBuffer.invert(np.arange(8)) # invert first sync byte of every group of eight 188-byte packets
        
        """
        ======================================================
        Scrambling/Energy Dispersal
        ======================================================
        """
        rand_seed = BitStream(bin = '100101010000000') # random seed for the pseudo-random binary sequence (PRBS) generator
        xor_out = rand_seed[13:14] ^ rand_seed[14:15] # start prbs on first bit at same time creates new BitStream object
        rand_seed >>= 1 # shift rand_seed to the right by 1 bit -- left most bit automatically set to 0
        rand_seed.set(xor_out,0) # set left most bit to xor_out
        
        # Now start loop to do the rest of the buffer. Since BitStream object now exists we can just append each bit.
        # It is important to keep the prbs output MSB on left and aligned with first bit of file stream 
        # also on left so we use append opposed to prepend. The first bit at the output of the generator is applied to
        # the first bit AFTER the first sync byte. The random generator never stops but the output is disabled for
        # the sync bytes thus they remain unscrambled. So, the size of the prbs generated data is 1503 bytes.
        # That is 8 frame * 188 bytes/frame - 1st sync byte = 1503 bytes. However, only 1496 bytes are scrambled, i.e.,
        # 8 frames * 188 data bytes/frame - 8 sync bytes.
        
        for i in np.arange(bufferSize):
            xor_out.append(rand_seed[13:14] ^ rand_seed[14:15]) # append next prbs output to BitStream object xor_out
            rand_seed >>= 1 # shift right
            rand_seed.set(xor_out[i+1:i+2],0) # replace left most bit of rand_seed with new value of xor_out
        
        fsize = 188 # frame size in bytes
        fsizeb = fsize*8 # frame size in bits
        
        for i in np.arange(8):
            streamBuffer[i*fsizeb+8:(i+1)*fsizeb] ^= xor_out[i*fsizeb+8-8:(i+1)*fsizeb-8] # scramble everything except sync bytes
        
        byte1 = mpeg2_parse(streamBuffer,1)
        byte2 = mpeg2_parse(streamBuffer,2)
        byte3 = mpeg2_parse(streamBuffer,3)
        byte4 = mpeg2_parse(streamBuffer,4)
        byte5 = mpeg2_parse(streamBuffer,5)
        byte6 = mpeg2_parse(streamBuffer,6)
        byte7 = mpeg2_parse(streamBuffer,7)
        byte8 = mpeg2_parse(streamBuffer,8)

        print(byte1.sync)
        print(byte2.sync)
        print(byte3.sync)
        print(byte4.sync)
        print(byte5.sync)
        print(byte6.sync)
        print(byte7.sync)
        print(byte8.sync)
        
        """
        ======================================================
        Create I/Q streams
        ======================================================
        """
    
        iData, qData = stream_to_QPSK(streamBuffer, upsample_rate) # Split bit stream into I and Q streams
        
        # Plot bits
        if plots == True:
            plt.title('iData stream')
            plt.plot(iData)
            plt.xlim(0,20*upsample_rate)
            plt.show()
        
        # Filter I and Q streams using convolution with RRC filter impulse response
        iModBits=np.convolve(h_rrc, iData, 'same')
        qModBits=np.convolve(h_rrc, qData, 'same')
        
        # Plot Filtered I and Q streams
        if plots == True:
            plt.title('iData filtered')
            plt.plot(iModBits)
            plt.xlim(0,20*upsample_rate)
            plt.show()
            plt.title('qData filtered')
            plt.plot(qModBits)
            plt.xlim(0,20*upsample_rate)
            plt.show()
        
        ### Perform modulation and Generate RF Output ###
        
        iMix = iMod*iModBits # modulate data with carrier
        qMix = qMod*qModBits # modulate data with carrier
        RfOut = iMix+qMix # sum I and Q signals to form RF signal
        
        ### Plot RF signal ###
        if plots == True:
            plt.title('RfOut and iMod')
            plt.plot(x, RfOut)
            plt.plot(x, iMod)
            plt.xlim(0,3*Tsymbol)
            plt.show()
        
        ### Plot RF PSD ###
        if plots == True:
            #Y = abs(np.fft.fft(RfOut))**2/len(x)/fSample
            Y = abs(np.fft.fft(RfOut))
            Y=10*np.log10(2*Y/Y.shape[-1])
            freq = np.fft.fftfreq(Y.shape[-1],1/fSample) # generates frequency bins in proper order (0,1.. n/2-1, -n/2... -1)/n normalized to +/-fs/2
            plt.title('RfOut PSD')
            plt.plot(freq,Y)
            plt.xlim(0,fSample/2) # plot one side of frequencies
            plt.show()
        
        # Plot RFOut Frequency Response
        if plots == True:
            Y = 2*abs(np.fft.fft(RfOut))/x.shape[-1] # x2 for one sided (SSB)
            freq = np.fft.fftfreq(x.shape[-1]) # 
            plt.title('RfOut Frequency Response')
            plt.plot(freq,Y)
            #plt.xlim(0,fMod/2) # plot one side of frequencies
            plt.show()

        """
        ======================================================
        SEND THROUGH CHANNEL AND ADD NOISE
        ======================================================
        """
        RfIn = awgn(RfOut, SNR_MIN, rate=1.0)
        
        """
        ======================================================
        RECEIVER SECTION
        ======================================================
        """

        ### Receive and Demodulate Signal
        #RfIn = RfOut
        iMixIn = iMod*RfIn
        qMixIn = qMod*RfIn
        
        # Plot PSD of De-modulated Signals
        if plots == True:
            Y = np.fft.fft(iMixIn)
            freq = np.fft.fftfreq(Y.shape[-1],1/fSample) # generates frequency bins in proper order (0,1.. n/2-1, -n/2... -1)/n normalized to +/-fs/2
            plt.plot(freq,10*np.log10(abs(Y)))
            plt.title('PSD of Received & Demodulated iData')
            plt.xlim(0,fSample/2) # plot one side of frequencies
            plt.show()
        
        if plots == True:
            Y = np.fft.fft(qMixIn)
            freq = np.fft.fftfreq(Y.shape[-1],1/fSample) # generates frequency bins in proper order (0,1.. n/2-1, -n/2... -1)/n normalized to +/-fs/2
            plt.plot(freq,10*np.log10(abs(Y)))
            plt.title('PSD of Received & Demodulated qData')
            plt.xlim(0,fSample/2) # plot one side of frequencies
            plt.show()
        
        # Filter De-modulated Signals
        iModBitsIn = np.convolve(h_rrc, iMixIn, 'same')
        qModBitsIn = np.convolve(h_rrc, qMixIn, 'same')
        
        # Plot PSD of De-modulated and Filtered Signals
        if plots == True:
            Y = np.fft.fft(iModBitsIn)
            freq = np.fft.fftfreq(Y.shape[-1],1/fSample) # generates frequency bins in proper order (0,1.. n/2-1, -n/2... -1)/n normalized to +/-fs/2
            plt.plot(freq,10*np.log10(abs(Y)))
            plt.title('PSD of Received, Demodulated and Filtered iData')
            plt.xlim(0,fSample/2) # plot one side of frequencies
            plt.show()
            
        # Plot Filtered Signals
        if plots == True:
            plt.plot(iModBitsIn)
            plt.plot(iData)
            plt.axvline((-0.5 + 4)*upsample_rate)
            plt.xlim(0,20*upsample_rate)
            plt.title('iData Symbols')
            plt.show()
            plt.plot(qModBitsIn)
            plt.plot(qData)
            plt.xlim(0,20*upsample_rate)
            plt.title('qData Symbols')
            plt.show()
             
        # Convert received analog back to digital bits
        iDataIn = 0.5*(np.sign(iModBitsIn[0.5*upsample_rate::upsample_rate])+1) # Detect state at symbol null periods
        qDataIn = 0.5*(np.sign(qModBitsIn[0.5*upsample_rate::upsample_rate])+1) # and convert to 1 or 0
        DataIn = np.vstack((iDataIn.astype(int),qDataIn.astype(int))).reshape((-1,),order='F') # Reassemble bit stream
        ErrVec = np.bitwise_xor(DataIn,streamBuffer[0:bufferSize]) # Create error vector based on sent vs received data
        all_zeros = not np.any(ErrVec) # Detect if any errors were detected
        print('No Errors = ', all_zeros)
        
        """
        ======================================================
        Write back to file
        ======================================================
        """
        
        # Write to file
        f = open('newfile.ts', 'ab')
        BitStream(DataIn).tofile(f)
        f.close()
        inc += 1
        #s = BitStream(filename='newfile.ts') # create BitStream object
        #test=s.read(bufferSize)
        #print(mpeg2_parse(test).payload[-10:])
        #print(frame.payload[-10:])

        """
        ======================================================
        Set up Plots
        ======================================================
        
        
        # definitions for the plot axes
        left, width=0.1,0.65
        bottom, height=0.1,0.65
        bottom_h=left_h=left+width+0.02
        
        rect_scatter=[left,bottom,width,height]
        rect_histx=[left,bottom_h,width,0.2]
        rect_histy=[left_h,bottom,0.2,height]
        
        # start with a rectangular figure
        plt.figure(1, figsize=(8,8))
        
        # set up plots
        axScatter=plt.axes(rect_scatter)
        axHistx=plt.axes(rect_histx)
        axHisty=plt.axes(rect_histy)
        
        # no axis labels for box plots
        axHistx.xaxis.set_major_formatter(NullFormatter())
        axHisty.yaxis.set_major_formatter(NullFormatter())
        
        
        ======================================================
        Generate scatter plots
        ======================================================
        
    
        # scatter plot:
        axScatter.scatter(numpy.real(w),numpy.imag(w))
        
        # draw axes at origin
        axScatter.axhline(0, color='black')
        axScatter.axvline(0, color='black')
        
        # add title (at x-axis) to scatter plot
        #title = 'Zero noise'
        title = 'SNR = %sdB with a BER of %s' % (SNR,BER)
        axScatter.xaxis.set_label_text(title)
        
        # now determine nice limits by hand:
        binwidth = 0.25 # width of histrogram 'bins'
        xymax = numpy.max( [numpy.max(numpy.fabs(numpy.real(w))), numpy.max(numpy.fabs(numpy.imag(w)))] ) # find abs max symbol value; nominally 1 
        lim = ( int(xymax/binwidth) + 1) * binwidth # create limit that is one 'binwidth' greater than 'xymax'
        
        axScatter.set_xlim( (-lim, lim) ) # set the data limits for the xaxis -- autoscale
        axScatter.set_ylim( (-lim, lim) ) # set the data limits for the yaxis -- autoscale
        
        bins = numpy.arange(-lim, lim + binwidth, binwidth) # create bins 'binwidth' apart between -lin and +lim -- autoscale
        axHistx.hist(numpy.real(w), bins=bins) # plot a histogram - xaxis are real values
        axHisty.hist(numpy.imag(w), bins=bins, orientation='horizontal') # plot a histogram - yaxis are imaginary values
        
        axHistx.set_xlim( axScatter.get_xlim() ) # set histogram axes to match scatter plot axes limits
        axHisty.set_ylim( axScatter.get_ylim() ) # set histogram axes to match scatter plot axes limits
        """

except ReadError: # read file until it's empty
    print("EOF")
    pass
```
